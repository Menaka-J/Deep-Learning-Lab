**AIM:**
    To implement and verify basic logic functions (AND, OR, NOT, NAND, NOR) using the McCullochâ€“Pitts (MP) neuron model.

**McCullouch Pits Neuron**

```python
def mcp_neuron(excitatory,inhibitory,threshold):
    if any(inhibitory):
        return 0
    return 1 if sum(excitatory)>=threshold else 0
```

**AND GATE**

```python
print("AND GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", mcp_neuron([x1,x2,x3],[],3))
```

**OR GATE**

```python
print("OR GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", mcp_neuron([x1,x2,x3],[],1))
```

**NOT GATE**

```python
print("NOT GATE:")
for x in [0,1]:
    print((x), "->", mcp_neuron([],[x],0))
```

**NAND GATE**

```python
print("NAND GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            and_out = mcp_neuron([x1,x2,x3],[],3)
            print((x1,x2,x3), "->", mcp_neuron([],[and_out],0))
```

**NOR GATE**

```python
print("NOR GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", mcp_neuron([],[x1,x2,x3],0))
```

**RESULT:**
    The MP neuron model was successfully implemented to realize AND, OR, NOT, NAND, and NOR logic functions, and the obtained outputs matched the respective truth tables, thereby validating the correct functioning of each logic gate.
