**AIM:**
    To implement and verify basic logic functions (AND, OR, NOT, NAND, and NOR) using the Perceptron learning model.

**PERCEPTRON**

```python
def perceptron(inputs, weight, bias):
    net = sum(i*w for i,w in zip(inputs,weight))
    net += bias
    if net>=0:
        return 1
    else:
        return 0
```

**AND GATE**

```python
print("AND GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", perceptron([x1,x2,x3],[1,1,1],-2.5))
```

**OR GATE**

```python
print("OR GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", perceptron([x1,x2,x3],[1,1,1],-0.5))
```

**NOT GATE**

```python
print("NOT GATE:")
for x in [0,1]:
    print((x), "->", perceptron([x],[-1,],0.5))
```

**NAND GATE**

```python
print("NAND GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", perceptron([x1,x2,x3],[-1,-1,-1],2.5))
```

**NOR GATE**

```python
print("NOR GATE:")
for x1 in [0,1]:
    for x2 in [0,1]:
        for x3 in [0,1]:
            print((x1,x2,x3), "->", perceptron([x1,x2,x3],[-1,-1,-1],0.5))
```

**XOR GATE**

```python
print("XOR GATE:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        or_out = perceptron([x1, x2], [1, 1], -0.5)
        and_out = perceptron([x1, x2], [1, 1], -1.5)
        not_and = perceptron([and_out], [-1], 0.5)
        xor_out = perceptron([or_out, not_and], [1, 1], -1.5)

        print((x1, x2), "->", xor_out)
```

**RESULT:**
    The Perceptron model was successfully trained to implement AND, OR, NOT, NAND, and NOR logic functions, and the outputs obtained matched the respective truth tables, confirming correct learning and classification by the perceptron.
